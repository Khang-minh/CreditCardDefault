```python
import pandas as pd

df = pd.read_csv("CreditCardDefault.csv", header=1)
df = df.drop(df.columns[0], axis=1)
```

```python
df
```

```python
new_names = {
    'LIMIT_BAL': 'CreditAmount',
    'SEX': 'Gender',
    'EDUCATION': 'Education',
    'MARRIAGE': 'MaritalStatus',
    'AGE': 'Age',
}
df = df.rename(columns=new_names)
```

```python
df
```

```python
# Bar plot for default status
df['default payment next month'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Default Status')
plt.xlabel('Default Status')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not Default', 'Default'], rotation=0) 
plt.show()

# Box plot for amount of credit by default status
df.boxplot(column='CreditAmount', by='default payment next month', grid=False, patch_artist=True, medianprops={'color':'black'})
plt.title('Amount of Credit by Default Status')
plt.xlabel('Default Status')
plt.ylabel('Amount of Credit (in millions)')
plt.xticks([1, 2], ['Not Default', 'Default'], rotation=0) 
plt.show()

# Box plot for age by default status
df.boxplot(column='Age', by='default payment next month', grid=False, patch_artist=True, medianprops={'color':'black'})
plt.title('Age by Default Status')
plt.xlabel('Default Status')
plt.ylabel('Age')
plt.xticks([1, 2], ['Not Default', 'Default'], rotation=0) 
plt.show()

# Correlation matrix to visualize relationships between variables
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation='vertical')
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Create a cross-tabulation of default status and marital status
cross_tab = pd.crosstab(df['MaritalStatus'], df['default payment next month'])

# Plot a grouped bar plot
cross_tab.plot(kind='bar', stacked=False, color=['skyblue', 'salmon'])
plt.title('Default Status by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.legend(title='Default Status', labels=['Not Default', 'Default'])
plt.xticks([0, 1, 2, 3], ['NA','Married', 'Single', 'Other'], rotation=0) 
plt.show()
```

```python
# Create a cross-tabulation of default status and marital status
cross_tab1 = pd.crosstab(df['Education'], df['default payment next month'])

# Plot a grouped bar plot
cross_tab1.plot(kind='bar', stacked=False, color=['skyblue', 'salmon'])
plt.title('Default Status by Education')
plt.xlabel('Education')
plt.ylabel('Count')
plt.legend(title='Default Status', labels=['Not Default', 'Default'])
plt.xticks([1, 2, 3, 4], ['Graduate','University', 'High school', 'Others'], rotation=0) 
plt.show()
```
